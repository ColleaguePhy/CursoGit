### ¿Qué es el control de versiones? 
Es un sistema que registra los cambios en un archivos o un conjunto de archivos en el tiempo de manera que podamos obtener versiones específicas después.
* Permite regresar un archivo a su estado anterior.
* Revertir todo el proceso a un estado anterior.
* Comparar cambios en el tiempo.
* Ver quien modifico el archivo.

<center>
<img src="Imagenes/kermit.jpg" width=50%>
</center>
<h3><center> Es una forma de protegernos de nosotros mismos.</center></h3>

### Control de versiones - Local

Son bases de datos locales que guardan todos los archivos bajo control y permiten revertir el estado de estos a una versión anterior.

El problema es al momento de colaborar con otros desarrolladores.
<center>
<img src="Imagenes/local_vcs.png" width=50%>
</center>

### Control de versiones - Centralizado
Un servidor contiene todos los archivos en diferentes versiones y loss clientes sacan la última versión del archivo del sistema central. 

El problema es que al fallar el servidor central durante un tiempo, impide que se pueda colaborar o guardar nuevas versiones.

<center>
<img src="Imagenes/centralized_vcs.png" width=50%>
</center>

### Control de versiones - Distribuido
En esta versión los clientes no solo sacan la ultima versión, sino que copian todo el repsitorio incluyendo el historial de versiones. Cada copia es un respaldo de todos los datos y se puede usar para restaurarlo de ser necesario

<center>
<img src="Imagenes/distributed_vcs.png" width=50%>
</center>

# Control de versiones - Local

### Estados
Git tiene tres estados principales donde estarán los archivos: *committed*, *staged* y *modified*
* <span style="color:red">**Modificado**</span> (*modified*) cambia un archivo pero no lo hemos guardado en la base de datos.
* <span style="color:green">**Preparado**</span> (*staged*) marcamos un archivo modificado en su versiónactual para guardarlo en la base de datos.
* <span style="color:gray">**Confirmado**</span> (*committed*) los datos están guardados en la base de datos local

### Directorios 
<center>
<img src="Imagenes/3_states.png" width=50%>
</center>

* El **directorio de trabajo** es donde estoy trabajando en los archivos, en una versión especifica del proyecto.
* El **área de preparación** es un área virtual, que guarda la información de los cambios que el usuario pretende registrar como una nueva versión.
* El **Repositorio Git** es donde Git guarda los metadatos, en descir el registro de los cambios y las versiones del proyecto que se han decidido salvar.

### Ramas
A partir de una base común podemos crear ramas para cada nueva funcionalidad, la idea con separar el proyecto en múltiples ramas es encapsular los cambios y hace más difícil  que código inestable se mezcle con la rama principal.

<center>

</center>

### Merge
Un merge es la forma de unir la historia de ramas separadas e integrarlas en una sola rama. Git automáticamente une los commits a menos que existan conflictos en ambas ramas.

Utiliza tu nombre y dirección de correo en lugar de los de Drácula, estos se asociarán coon tu actividad posterior de Git, lo que significa que cualquier cambio realizado en los servidores de Git como GitHub o GitLab incluirá esta información

Para entender mejor las diferencias entre checkout, revert y reset haga clic [aquí](https://stackoverflow.com/questions/8358035/whats-the-difference-between-git-revert-checkout-and-reset)

## Ignorando cosas
¿Qué pasa si tenemos archivos que no queremos que Git rastree?

Vamos a crear algunos archivos ficticios en un nuevo directorio
``` bash
mkdir results
```
 
``` bash
touch a.dat b.dat c.dat results/a.out results/b.out
```
Miremos el estatus 
``` bash
git status
```
<center>
<img src="Imagenes/git_status_09.png" width=50%>
</center>

Colocar estos archivos bajo el control de versiones sería un desperdicio de espacio en disco. Y lo que es peor, al tenerlos todos listados, podría distraernos de los cambios que realmente importan, así que vamos a decirle a Git que los ignore.

Lo hacemos creando un archivo en el directorio raíz de nuestro proyecto llamado **.gitignore** y agregando los archivos que deseamos

``` bash
cat .gitignore
```
<center>
<img src="Imagenes/git_cat_05.png" width=10%>
</center>

Volvemos a mirar el estatus y vemos que ahora la salida es mucho más limpia

``` bash
git status
```
<center>
<img src="Imagenes/git_status_10.png" width=50%>
</center>

**Nota:** una vez Git empieza a seguir los archivos, lo seguira haciendo así los agregemos al .gitignore. Podemos agregar este archivo a nuestro repositorio si queremos compartirlo con nuestros colaboradores. 

Para colocar una excepción en el **.gitignore** puedes hacerlo de la siguiente forma
``` bash
.data  # Ignora todos los archivos .data
!final.data   # Excepto el archivo final.data
```

**Importante** Si queremos dejar de seguir algun archivo podemos usar la siguiente linea

``` bash
git rm --cached <file>
```

## Trabajando con GitHub
Github es una plataforma para control de versiones y **colaboración**. Permite que múltipes personas puedan trabajar en proyectos sin importar donde están.

* Múltiples empresas y universidades comparten sus desarrollos.
* Es totalmente gratis siempre y cuando el acceso sea público.
* Podemos subir cualquier tipo de elementos, no sólo código.
* Podemos conocer las mejores prácticas en cualquier lenguaje.

### Taller de Github

* Cree una cuenta en www.github.com


* (El ejercicio se hará en parejas, cada uno de los estudiantes debe asumir un rol **A** y **B**)


* El estudiante A debe crear un repositorio **VACIO** usando la plataforma.

#### Configurar el repositorio 
El estudiante **A** configura el remoto en su proyecto local usando:

```bash
git remote add origin <<URL_REPO>>
git push -u origin master
```

El estudiante **B** en el escritorio clonará el repositorio del estudiante *A*

```bash 
git clone <<URL_REPO>>
```

#### ¿Qué acabamos de hacer?

Para comunicarnos con Github debemos configurar el repositorio externo, un **remoto** al que llamamos **origin** para que podamos subir los cambios.

```bash
git remote add origin <<URL_REPO>>
```
---

Este es un comando que dice *Suba todos los commits en la **rama** llamada master* al remoto llamado origin.

```bash
git push origin master
```
---

Es usado para apuntar a un repo existente y hacer una copia en un nuevo directorio. Esto copia toda la historia del repositorio, todos los commits y la **rama** a la que apunta el repo remoto.

```bash
git clone <<URL_REPO>>
```

#### Subiendo cambios y archivos


1. El estudiante **A** modifica el archivo `AUTHORS.md` agregando nuevos autores, hace COMMIT y sube los cambios a Github con

```bash
git push
```

3. El estudiante **B** obtiene los cambios que A subió a Github.

```bash
git pull
```

4. Ahora **B** ya puede subir los cambios que había creado a Github.

5. El estudiante **A** se asegura de obtener los cambios de B en local.

Este comando se usa para subir el contenido de un repositorio local a un repositorio remoto. Se transfieren todos los commits.

```bash
git push
```

Este comando es usado para descargar el contenido de un repositorio remotoe inmediatamente **mezclar** con el repositorio local para sincronizar el contenido.

```bash
git pull
```
